```python
diseases = ['164889003', '195080001', '164890007', '427393009', '426627000', '270492004']
def find_disease(filename):
    lines = []
    with open(filename) as f:
        lines = f.readlines()
    line = lines[15]
    mil_gai = False
    for d in diseases:
        if d in line:
            mil_gai = True
            break
    if mil_gai:
        return d, filename
    return None
```

```python
os.listdir("./training")
```

```python
for root, folders, files in os.walk('./training/cpsc_2018/'):
    for file in files:
        if file.endswith('.hea'):
            temp = find_disease(f"{root}/{file}")
            if temp:
                print(temp)
```
